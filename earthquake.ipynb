```python
# Extract latitude, longitude, and magnitude columns
fig, ax=plt.subplots()
latitude = eqData['latitude']
longitude = eqData['longitude']
magnitude = eqData['mag']
# Scatter plot earthquakes on the map
sc = ax.scatter(longitude, latitude, c=magnitude, cmap='viridis', marker='o', s=magnitude * 10, alpha=0.7)
plt.colorbar(sc, label='Magnitude')
fig.tight_layout()
```

```python
import cartopy.crs as ccrs

# Create a basic map using Cartopy
plt.figure(figsize=(10, 6))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()
# Scatter plot earthquakes on the map, the *10 is the size of magnitude bubble
sc = ax.scatter(eqData['longitude'], eqData['latitude'], c=eqData['mag'], cmap='viridis', marker='o', s=eqData['mag'] * 10, alpha=0.7)
# Add colorbar
plt.colorbar(sc, label='Magnitude')
# Set plot title
plt.title('Earthquake Distribution Map')
# Show the plot
plt.show()
fig.tight_layout()
```

```python
# cretae correlation matrix
# in heatmap with correlation matrix (dont need to define variables) 

corr_mat = eqData.corr(numeric_only=True)
corr_mat
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(corr_mat, annot=True)
```

```python
# create a subset of the data of to see the correlation matrix of the specified variables and drop any missing variables

magdepth_df = eqData[['mag', 'depth', 'nst']].dropna()
corr_magdepth = magdepth_df.corr()
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(corr_magdepth, annot=True)
```

```python
# scatterplot figure and all the markers are blue
plt.figure(figsize=(8, 6))
plt.scatter(eqData['nst'], eqData['mag'], c='blue', alpha=0.7)

# Add labels and title
plt.xlabel('Number of Stations')
plt.ylabel('Magnitude')
plt.title('Scatter Plot of Number of Stations vs Magnitude')

# Show plot
plt.grid(True)
plt.show()
```

```python


# Filter the dataframe to exclude data points with number of stations > 300
eq_df = pd.DataFrame(eqData)
eq_df_filtered = eq_df[eq_df['nst'] <= 300]

# Create a scatter plot of the number of stations and magnitude and the markers will be assigned a colour of the colorbar
plt.figure(figsize=(8, 6))
plt.scatter(eq_df_filtered['nst'], eq_df_filtered['mag'], c=eq_df_filtered['mag'], cmap='viridis', alpha=0.7)

# Add colorbar
plt.colorbar(label='Magnitude')


# Add labels and title
plt.xlabel('Number of Stations')
plt.ylabel('Magnitude')
plt.title('Scatter Plot of Number of Stations vs Magnitude (nst <= 300)')
```
